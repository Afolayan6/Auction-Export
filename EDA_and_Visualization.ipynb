```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("Auction_export_cleaned")
df = df.drop(columns = "Unnamed: 0")
```

### The correlation betweeen the continuous values

```python
df_corr_1 = df.drop(columns = ["Brand", "Model", "State", "Color", "Body"])
```

```python
df_corr_2 = df_corr_1.corr()
```

```python
plt.figure(figsize = (10, 5))
sns.heatmap(df_corr_2)
plt.show()
```

### Relationship between Year and Price

```python
df_1 = df.drop(columns = ["Brand", "Model", "State", "Milage", "Color", "Body"])
```

```python
df_years = df_1.groupby("Year").mean().reset_index()
```

```python
plt.figure(figsize = (10, 5))
plt.plot(df_years.Year, df_years.Price, color = "green", lw = 4.5, ls = "-")
plt.scatter(df.Year, df.Price)
plt.xlabel("Year")
plt.ylabel("Price")
plt.show()
```

### Relationship between Milage and Price

```python
plt.scatter(df.Milage, df.Price)
plt.xlabel("Milage")
plt.ylabel("Price")
plt.show()
```

```python
#To seperate the outliers
df_filt = df[df["Milage"] < 800000]
```

```python
plt.figure(figsize = (10, 5))
plt.scatter(df_filt.Milage, df_filt.Price)
plt.xlabel("Milage")
plt.ylabel("Price")
plt.show()
```

### Comparing prices for different brands

```python
df_2 = df.drop(columns = ["Year", "Model", "State", "Milage", "Color", "Body"])
df_brands = df_2.groupby("Brand").mean().reset_index().sort_values(by = "Price")
```

```python
plt.figure(figsize=(14, 6))
sns.barplot(x = "Brand", y = "Price", data = df_brands)
plt.show()
```

### Distribution of the body types

```python
plt.figure(figsize = (10, 5))
sns.histplot(x = "Body", data = df)
plt.show()
```
